```python
import requests
import pandas as pd
import csv
import datetime
```

# Colombia

```python
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'
parameters = {
    'format': 'geojson',
    'starttime': '1990-01-01',
    'endtime': '2023-06-28',
    'minmagnitude': '2.5',
    'minlatitude': '-5.266',
    'maxlatitude': '15.708',
    'minlongitude': '276.328',
    'maxlongitude': '293.906',
    'limit': 20000
}

response = requests.get(url, params=parameters)
data = response.json()

earthquakes = data['features']
records = []

for eq in earthquakes:
    latitude = eq['geometry']['coordinates'][1]
    longitude = eq['geometry']['coordinates'][0]
    magnitude = eq['properties']['mag']
    intensity = eq['properties']['cdi'] 
    date = pd.to_datetime(eq['properties']['time'], unit='ms')
    location = eq['properties']['place']
    
    records.append({
        'Fecha': date,
        'Latitud': latitude,
        'Longitud': longitude,
        'Magnitud': magnitude,
        'Intensidad': intensity,
        'Ubicacion': location
    })

# Crear el DataFrame
df_Colombia = pd.DataFrame(records)
df_Colombia = pd.concat([df_Colombia, pd.DataFrame(data['features']).properties.apply(pd.Series)],axis=1)
df_Colombia.info()
```
