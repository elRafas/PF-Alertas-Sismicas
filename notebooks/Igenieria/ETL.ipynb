# Colombia

```python
import requests
import pandas as pd
import csv
import datetime
import reverse_geocoder as rg
```

```python
# URL de la API
url = 'https://earthquake.usgs.gov/fdsnws/event/1/query'

# Parámetros de la consulta
params = {
    'format': 'geojson',
    'starttime': '1990-01-01',
    'endtime': '2023-06-28',
    'minmagnitude': '2.5',
    'minlatitude': '-5.266',
    'maxlatitude': '15.708',
    'minlongitude': '276.328',
    'maxlongitude': '293.906',
    'limit': 20000
}

# Realizar la solicitud GET a la API
response = requests.get(url, params=params)

# Obtener los datos en formato JSON
data = response.json()
# Crear una lista vacía para almacenar los registros
records = []

# Recorrer los eventos y extraer los parámetros
for event in data['features']:
    properties = event['properties']
    latitude = event['geometry']['coordinates'][1]
    longitude = event['geometry']['coordinates'][0]
    longitude2 = event['geometry']['coordinates'][2]
    
    

    records.append(properties)

for event in data['features']:
    properties = event['properties']
    coordinates = event['geometry']['coordinates']
    properties['latitude'] = coordinates[1]
    properties['longitude'] = coordinates[0]
    properties['hipocentro'] = coordinates[2]
    
    
    records.append(properties)

# Crear el DataFrame
df_Colombia = pd.DataFrame(records)

df_Colombia.info()

```

```python
# Damos formato de fecha a la columna "time"
df_sismos["time"] = pd.to_datetime(df_sismos["time"], unit= "ms")
 

# Damos formato de fecha a la columna "time"
df_sismos["updated"] = pd.to_datetime(df_sismos["updated"], unit= "ms")
```

```python
# Quitemos los nulos
df_sismos = df_sismos[["time", "mag", "cdi", "gap", "latitude", "longitude",  "place", "ids","hipocentro"]]
df_sismos = df_sismos.dropna()
df_sismos.info()
df_sismos.head(3)
```

```python
# Creamos las columnas con "Pais" y con "Region"

# Crear una lista de tuplas de coordenadas (latitud, longitud)
coordinates = list(zip(df_sismos['latitude'], df_sismos['longitude']))

# Realizar la búsqueda inversa de las coordenadas
results = rg.search(coordinates)

# Obtener la lista de prefecturas a partir de los resultados
prefectures = [result['admin1'] if result['admin1'] != '' else 'Desconocido' for result in results]
countries = [result['cc'] if result['cc'] != '' else 'Desconocido' for result in results]
# Agregar la columna "region" al DataFrame
df_sismos['Region'] = prefectures
df_sismos['Pais'] = countries
df_sismos.info()
df_sismos.head(5)
```

```python
# Listado de valores a eliminar en la columna 'Col1'
paises_permitidos = ['CO', 'US', 'JP']

# Eliminar las filas donde los valores de 'Pais' no coinciden con el listado
df_sismos = df_sismos[df_sismos['Pais'].isin(paises_permitidos)]

# Renombramos las columnas
nuevos_nombres = {'time': "Fecha", 'mag': "Magnitud", 'cdi': "Intensidad", 'gap': "GAP", 'latitude': "Latitud", 'longitude': "Longitud", 'place': "Lugar", 'ids': "Id", "hipocentro": "Hipocentro", 'Region': 'Region' , 'Pais': 'Pais'}
df_sismos.columns = nuevos_nombres.values()


# Creamos un csv df_completo

#df_sismos.to_csv("../../../DataSet/data_sismos_normalizado.csv")
df_sismos.to_csv("../../DataSet/data_sismos_normalizado.csv")


# Mostrar el DataFrame resultante
df_sismos.info()
df_sismos.head(5)
```
